```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("Kabir_singh.csv")
data2 = pd.read_csv("Arjun_reddy.csv")
```

```python

def calculate_mean_and_median(data):
    frequency = data["Frequency"]
    no_of_observ = sum(frequency)

    xi = data["Rating"]
    fixi = [xi * freq for xi, freq in zip(xi, frequency)]
    mean = sum(fixi) / no_of_observ

    values = []
    for rating, frequency in data[['Rating', 'Frequency']].values:
        values.extend([rating] * int(frequency))

    values.sort()
    median_index = len(values) // 2
    median = values[median_index] if len(values) % 2 == 1 else (values[median_index - 1] + values[median_index]) / 2

    return mean, median
```

```python

Ac1 = data["Label"].to_list()

mean_kabir, median_kabir = calculate_mean_and_median(data)
print("Kabir Singh - Mean:", mean_kabir,"||" ,"Median:", median_kabir,"||","Actual Rating : ",Ac1[0])

Ac2 = data2["Labels"].to_list()

mean_arjun, median_arjun = calculate_mean_and_median(data2)
print("Arjun Reddy - Mean:", mean_arjun,"||" ,"Median:", median_arjun,"||","Actual Rating : ",Ac2[0])
```

```python

labels = ['Mean', 'Median', 'Actual']
kabir_values = [mean_kabir, median_kabir, Ac1[0]]
arjun_values = [mean_arjun, median_arjun, Ac2[0]]

x = range(len(labels))

plt.bar(x, kabir_values, width=0.4, label='Kabir Singh')
plt.bar([i + 0.4 for i in x], arjun_values, width=0.4, label='Arjun Reddy')

plt.xlabel('Metrics')
plt.ylabel('Values')
plt.title('Comparison of Mean, Median, and Actual Ratings')
plt.xticks([i + 0.2 for i in x], labels)
plt.legend()

plt.tight_layout()
plt.show()
```
